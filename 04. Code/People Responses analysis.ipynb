```python
import numpy as np
import pandas as pd

# Cleaning and finding the mean percent change from baseline for each category
# However, due to limited number of countries with available data, it is not feasable to draw any conclusions from this dataset

mobility_df = pd.read_csv("Global_Mobility_Report.csv")
mobility_df['retail_and_recreation_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['retail_and_recreation_percent_change_from_baseline'].transform('mean')
mobility_df['grocery_and_pharmacy_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['grocery_and_pharmacy_percent_change_from_baseline'].transform('mean')
mobility_df['parks_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['parks_percent_change_from_baseline'].transform('mean')
mobility_df['transit_stations_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['transit_stations_percent_change_from_baseline'].transform('mean')
mobility_df['workplaces_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['workplaces_percent_change_from_baseline'].transform('mean')
mobility_df['residential_percent_change_from_baseline'] = mobility_df.groupby(['country_region'])['residential_percent_change_from_baseline'].transform('mean')

mobility_df = mobility_df[['country_region', 'retail_and_recreation_percent_change_from_baseline'
                          , 'grocery_and_pharmacy_percent_change_from_baseline',
                          'parks_percent_change_from_baseline',
                          'transit_stations_percent_change_from_baseline',
                          'workplaces_percent_change_from_baseline',
                          'residential_percent_change_from_baseline']].drop_duplicates()
mobility_df.reset_index()
```
